```python
import pandas as pd
import numpy as np
from keras.preprocessing.sequence import pad_sequences
```

```python
data_df = pd.read_csv("Data/data_with_industry_codes.csv")
questions_df = pd.read_csv("Data/questions.csv")
questions_df = questions_df.filter(['questions_id','questions_title','questions_date_added','questions_body'])
best_answer_df = pd.merge(data_df, questions_df, on="questions_id")
best_answer_df = best_answer_df.drop(columns=['Unnamed: 0'])
# best_answer_df_short = best_answer_df.head(10)
# best_answer_df_short

best_answer_df
```

```python
from cypher import encode
from datetime import datetime
```

```python
best_answer_df['indexed'] = best_answer_df.apply(lambda x: encode(x['questions_title'], x['questions_body']), axis=1)
```

```python
ba_df = best_answer_df
ba_df.dtypes
```

```python
ba_df['industry_codes'] = ba_df['industry_codes'].astype(int)
ba_df.dtypes
```

```python
indexed_padded = pad_sequences(best_answer_df['indexed'], value=0, padding='post', maxlen=50)
```

```python
indexed_padded[666]
```

```python
import keras
import tensorflow as tf
```

```python
from sklearn.model_selection import train_test_split

X=indexed_padded
y = best_answer_df["industry_codes"]

print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from tensorflow.keras.utils import to_categorical
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=42)
```

```python
X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
X_test_scaled[100]
```

```python
# Step 1: Label-encode data set
label_encoder = LabelEncoder()
label_encoder.fit(y_train)
# label_encoder.fit(y_test)
encoded_y_train = label_encoder.transform(y_train)
# encoded_y_test = label_encoder.transform(y_test)
encoded_y_train
```

```python
# Step 2: Convert encoded labels to one-hot-encoding
y_train_categorical = to_categorical(encoded_y_train)
# y_test_categorical = to_categorical(encoded_y_test)

y_train_categorical
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten

# input shape is the vocabulary count used for the movie reviews (10,000 words)
# vocab_size = 10000
# model = keras.Sequential()
# model.add(keras.layers.Embedding(vocab_size, 50))
# model.add(keras.layers.GlobalAveragePooling1D())
# model.add(keras.layers.Dense(1, activation=tf.nn.relu))
# model.add(keras.layers.Dense(821, activation=tf.nn.relu))
# model.summary()
```

```python
# Create model and add layers
model = Sequential()
model.add(Dense(units=100, activation='relu', input_dim=50))
model.add(Dense(units=100, activation='relu'))
model.add(Dense(units=821, activation='sigmoid'))
```

```python
# Compile the model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# Fit the model to the training data
model.fit(
    X_train_scaled,
    encoded_y_train,
    epochs=1000,
    shuffle=True,
    verbose=1
)
```

```python
model_loss, model_accuracy = model.evaluate(
    X_train_scaled, encoded_y_train, verbose=2)
print(
    f"Normal Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Save the model
model.save("first_model_50.h5")
```
