```python
from cypher import encode
from datetime import datetime
```

```python
for i in np.arange(0,len(best_answer_df_short)):
    best_answer_df_short.iloc[i]['indexed'] = encode(str(best_answer_df_short["questions_title"]), str(best_answer_df_short["questions_body"]))
```

```python
best_answer_df['indexed'] = best_answer_df.apply(lambda x: encode(x['questions_title'], x['questions_body']), axis=1)
```
