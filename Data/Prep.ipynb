```python
import pandas as pd
```

```python
a_df = pd.read_csv("answers.csv")
p_df = pd.read_csv("professionals.csv")
q_df = pd.read_csv("questions.csv")

a_df.rename(columns={'answers_author_id':'professionals_id',
                          'answers_question_id':'questions_id'}, 
                 inplace=True)
```

```python
a_df.head()
```

```python
p_df.head()
```

```python
result = pd.merge(a_df, p_df, on="professionals_id")
result.shape
```

```python
result.head()
```

```python
result_df = pd.merge(result, q_df, on="questions_id")
```

```python
result_df.head()
```

```python
result_df["professionals_industry"].nunique()
```

```python
result_df.shape
```

```python
result_df["professionals_industry"].head()
```

```python
result_df.dropna(subset=["professionals_industry"], inplace = True)
```

```python
result_df.head()
```

```python
result_df2 = result_df.filter(['professionals_industry','questions_body', 'questions_id'])
```

```python
result_df2.head()
result_df2.shape
```

```python
result_df2['questions_body'] = result_df2['questions_body'].str.replace('#','')
result_df2['questions_body'] = result_df2['questions_body'].str.replace('\r\n','')
```

```python
result_df2.to_csv ('data.csv', encoding="utf-8", index = False, sep=',')
```
