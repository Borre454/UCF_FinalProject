```python
import pandas as pd
import numpy as np
```

```python
rank = pd.read_csv('Data/questions.csv')
rank.head()
```

```python
#for line in np.arange(0, len(rank['questions_title'])):
for word in rank.iloc[0]['questions_title'].split():
        print(word)
```

```python
counts = {}
for line in np.arange(0, len(rank['questions_title'])):
    for word in rank.iloc[line]['questions_title'].split():
        if word in counts:
            key = word.lower()
            key = key.strip('?.!:')
            counts[key] += 1
        else:
            key = word.lower()
            key = key.strip('?.!:')
            counts[key] = 1
    for word in rank.iloc[line]['questions_body'].split():
        if word in counts:
            key = word.lower()
            key = key.strip('?.!:')
            counts[key] += 1
        else:
            key = word.lower()
            key = key.strip('?.!:')
            counts[key] = 1
```

```python
counts = sorted(counts.items(), key=lambda item: item[1])
```

```python
print(counts)
```

```python
indexed = {0: '<NULL>', 1:'<START>'}
index = 2
for key, value in counts:
    indexed[index] = key
    index += 1
```

```python
indexed = pd.DataFrame.from_dict(indexed, orient = 'index')
```

```python
indexed.to_csv('word_index.csv')
```
